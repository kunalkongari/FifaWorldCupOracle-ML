# 04 — Monte Carlo Tournament Simulation
**FIFA 2026 World Cup Predictor**

Uses the trained XGBoost + ELO model to simulate the entire 2026 World Cup **10000 times**.  
Each simulation samples match outcomes from predicted probabilities, giving us a full  
probability distribution over every team's tournament path.

**What this notebook produces:**
- Win probability per team (% of simulations they won the tournament)
- Final / Semi-final / Quarter-final appearance probabilities
- `outputs/monte_carlo_results.csv`
- Rich visualisations of the results


## Imports & load model artefacts

```python
import pandas as pd
import numpy as np
import pickle
import json
import warnings
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.stats import poisson
from collections import Counter

warnings.filterwarnings("ignore")
np.random.seed(42)

# Load trained artefacts from notebook 03
model      = pickle.load(open("outputs/model.pkl",      "rb"))
elo_dict   = pickle.load(open("outputs/elo_dict.pkl",   "rb"))
team_stats = pickle.load(open("outputs/team_stats.pkl", "rb"))
dict_table = pickle.load(open("data/dict_table_2026",   "rb"))
df_fixture = pd.read_csv("data/fixture_2026.csv")

print(f"Model loaded: {type(model).__name__}")
print(f"ELO ratings: {len(elo_dict)} teams")
print(f"Fixture: {len(df_fixture)} matches")
print(f"Groups: {list(dict_table.keys())}")
```

## Configuration

```python
N_SIMS = 10000

INITIAL_ELO = 1500

# UEFA/IC playoff path winners (approximated)
PATH_MAP = {
    "UEFA Path A winner": "Slovakia",
    "UEFA Path B winner": "Poland",
    "UEFA Path C winner": "Greece",
    "UEFA Path D winner": "Romania",
    "IC Path 1 winner":   "United States",
    "IC Path 2 winner":   "New Caledonia",
}

def clean(name: str) -> str:
    return str(name).replace("(H)", "").replace("[a]", "").strip()

print(f"Simulations to run: {N_SIMS:,}")
print(f"Path placeholders: {PATH_MAP}")
```

## Core simulation functions

```python
def get_probs(home, away, cache):
    """Return (p_home_win, p_draw, p_away_win) — uses cache for speed."""
    key = (home, away)
    if key in cache:
        return cache[key]

    h_elo = elo_dict.get(home, INITIAL_ELO)
    a_elo = elo_dict.get(away, INITIAL_ELO)
    h_h   = team_stats.get(home, [])[-10:]
    a_h   = team_stats.get(away, [])[-10:]
    hfs   = np.mean([x[0] for x in h_h]) if h_h else 1.5
    hfc   = np.mean([x[1] for x in h_h]) if h_h else 1.2
    afs   = np.mean([x[0] for x in a_h]) if a_h else 1.5
    afc   = np.mean([x[1] for x in a_h]) if a_h else 1.2

    lh, la = max(hfs * afc, 0.3), max(afs * hfc, 0.3)
    ph = pd_ = pa = 0.0
    for x in range(9):
        for y_s in range(9):
            p = poisson.pmf(x, lh) * poisson.pmf(y_s, la)
            if x > y_s:    ph  += p
            elif x == y_s: pd_ += p
            else:           pa  += p

    x_feat = np.array([[h_elo, a_elo, h_elo-a_elo, hfs, hfc, afs, afc, 0.33, 0.33]])
    ml     = model.predict_proba(x_feat)[0]

    p_h = 0.7*ml[2] + 0.3*ph
    p_d = 0.7*ml[1] + 0.3*pd_
    p_a = 0.7*ml[0] + 0.3*pa
    t   = p_h + p_d + p_a
    result = (p_h/t, p_d/t, p_a/t)
    cache[key] = result
    cache[(away, home)] = (result[2], result[1], result[0])
    return result


def sample_result(home, away, cache, knockout=False):
    """Sample a match result from predicted probabilities."""
    ph, pd_, pa = get_probs(home, away, cache)
    r = np.random.random()
    if knockout:
        # In knockout: draws go to extra time/penalties — equal coin flip
        return home if r < ph + pd_/2 else away
    else:
        if r < ph:        return "H"
        elif r < ph+pd_:  return "D"
        else:             return "A"


def simulate_once(cache):
    """Simulate the full 2026 World Cup. Returns (champion, group_tables, final_df, sf_df, qf_df)."""
    tables = {g: df.copy() for g, df in dict_table.items()}
    for g in tables:
        tables[g]["Team"] = tables[g]["Team"].apply(clean)
        tables[g]["Pts"]  = 0.0

    fix = df_fixture.copy()
    for col in ["Home", "Away"]:
        fix[col] = fix[col].apply(clean)
        for ph, team in PATH_MAP.items():
            fix[col] = fix[col].replace(clean(ph), team)

    df_group = fix.iloc[:72].copy()
    df_ko1   = fix.iloc[72:88].copy()
    df_ko2   = fix.iloc[88:96].copy()
    df_qf    = fix.iloc[96:100].copy()
    df_sf    = fix.iloc[100:102].copy()
    df_final = fix.iloc[103:].copy()

    # ── Group stage ────────────────────────────────────────────────────────────
    for grp, tbl in tables.items():
        grp_teams = set(tbl["Team"].values)
        for _, row in df_group[df_group["Home"].isin(grp_teams)].iterrows():
            h, a = row["Home"], row["Away"]
            res  = sample_result(h, a, cache)
            if   res=="H": tbl.loc[tbl["Team"]==h, "Pts"] += 3
            elif res=="D":
                tbl.loc[tbl["Team"]==h, "Pts"] += 1
                tbl.loc[tbl["Team"]==a, "Pts"] += 1
            else:           tbl.loc[tbl["Team"]==a, "Pts"] += 3
        tables[grp] = tbl.sort_values("Pts", ascending=False).reset_index(drop=True)

    thirds = sorted(
        [(g, tables[g].iloc[2]["Team"], tables[g].iloc[2]["Pts"]) for g in tables],
        key=lambda x: -x[2]
    )
    third_names = [t[1] for t in thirds]

    reps = {}
    for g in tables:
        l = g.split()[-1]
        reps[f"Winner Group {l}"]    = tables[g].iloc[0]["Team"]
        reps[f"Runner-up Group {l}"] = tables[g].iloc[1]["Team"]
        reps[f"3rd Group {l}"]       = tables[g].iloc[2]["Team"]

    slot_keys = [
        "3rd Group A/B/C/D/F","3rd Group C/D/F/G/H","3rd Group C/E/F/H/I",
        "3rd Group E/H/I/J/K","3rd Group A/E/H/I/J","3rd Group B/E/F/I/J",
        "3rd Group E/F/G/I/J","3rd Group D/E/I/J/L",
    ]
    for i, k in enumerate(slot_keys):
        reps[k] = third_names[i] if i < len(third_names) else "TBD"

    def fill(df):
        df = df.copy()
        for col in ["Home", "Away"]:
            df[col] = df[col].apply(lambda x: reps.get(clean(str(x)), clean(str(x))))
        return df

    def run_round(df):
        df = df.copy(); df["winner"] = ""
        for idx, row in df.iterrows():
            h, a = clean(str(row["Home"])), clean(str(row["Away"]))
            df.at[idx, "winner"] = (h if a=="TBD" else a) if "TBD" in [h,a]                                    else sample_result(h, a, cache, knockout=True)
        return df

    def advance(prev, nxt):
        nxt = nxt.copy()
        for _, row in prev.iterrows():
            mid = clean(str(row["Score"]))
            w   = clean(str(row["winner"]))
            for col in ["Home","Away"]:
                nxt[col] = nxt[col].apply(lambda x: w if clean(str(x))==f"Winner {mid}" else x)
        return nxt

    df_ko1   = run_round(fill(df_ko1))
    df_ko2   = run_round(advance(df_ko1, fill(df_ko2)))
    df_qf    = run_round(advance(df_ko2, fill(df_qf)))
    df_sf    = run_round(advance(df_qf,  fill(df_sf)))
    df_final = run_round(advance(df_sf,  fill(df_final)))

    champion = clean(str(df_final.iloc[0]["winner"])) if len(df_final) else "TBD"
    return champion, tables, df_final, df_sf, df_qf

print("Functions defined ✓")
```

## Pre-compute probability cache

```python
all_teams = set()
for g, tbl in dict_table.items():
    for t in tbl["Team"].apply(clean):
        all_teams.add(t)
all_teams.update(PATH_MAP.values())
all_teams.discard("TBD")

print(f"Building probability cache for {len(all_teams)} teams...")
prob_cache = {}
teams_list = list(all_teams)
for i, a in enumerate(teams_list):
    for b in teams_list[i+1:]:
        get_probs(a, b, prob_cache)

print(f"Cached {len(prob_cache)} matchup pairs — simulations will now be fast")
```

## Run Monte Carlo simulations

```python
wins        = Counter()
final_apps  = Counter()
semi_apps   = Counter()
qf_apps     = Counter()

print(f"Running {N_SIMS} simulations...")
for i in range(N_SIMS):
    champ, tables, df_f, df_s, df_q = simulate_once(prob_cache)
    wins[champ] += 1
    for _, r in df_f.iterrows():
        final_apps[r["Home"]] += 1;  final_apps[r["Away"]] += 1
    for _, r in df_s.iterrows():
        semi_apps[r["Home"]]  += 1;  semi_apps[r["Away"]]  += 1
    for _, r in df_q.iterrows():
        qf_apps[r["Home"]]    += 1;  qf_apps[r["Away"]]    += 1

print(f"Done! Champion distribution across {N_SIMS} sims:")
for team, count in wins.most_common(10):
    print(f"  {team:<22} {count:>4} wins  ({count/N_SIMS*100:.1f}%)")
```

## Build results table

```python
results = {}
for team in set(list(wins) + list(final_apps) + list(semi_apps)):
    if team and team != "TBD" and len(team) > 2:
        results[team] = {
            "win_pct":   round(wins[team]       / N_SIMS * 100, 1),
            "final_pct": round(final_apps[team] / N_SIMS * 100, 1),
            "semi_pct":  round(semi_apps[team]  / N_SIMS * 100, 1),
            "qf_pct":    round(qf_apps[team]    / N_SIMS * 100, 1),
        }
results = dict(sorted(results.items(), key=lambda x: -x[1]["win_pct"]))

mc_df = pd.DataFrame([{"Team": t, **s} for t, s in results.items()])
mc_df.to_csv("outputs/monte_carlo_results.csv", index=False)

with open("outputs/mc_results.json", "w") as f:
    json.dump(results, f, indent=2)

print("Saved: outputs/monte_carlo_results.csv")
print()
print(f"{'Team':<22}  {'Win%':>6}  {'Final%':>7}  {'Semi%':>7}  {'QF%':>5}")
print("-" * 56)
for team, s in list(results.items())[:20]:
    print(f"{team:<22}  {s['win_pct']:>5.1f}%  {s['final_pct']:>6.1f}%  {s['semi_pct']:>6.1f}%  {s['qf_pct']:>4.1f}%")
```

## Visualisation 1 — Win probability bar chart

```python
top15 = list(results.items())[:15]
teams_plot = [t for t,_ in top15]
win_vals   = [s["win_pct"] for _,s in top15]
colors_win = cm.YlOrRd(np.linspace(0.35, 0.9, len(top15)))[::-1]

fig, ax = plt.subplots(figsize=(11, 6))
bars = ax.barh(teams_plot[::-1], win_vals[::-1], color=colors_win, edgecolor="white", linewidth=0.5)
ax.bar_label(bars, labels=[f"{v:.1f}%" for v in win_vals[::-1]], padding=5, fontsize=9, fontweight="bold")
ax.set_title(f"FIFA 2026 World Cup — Win Probabilities ({N_SIMS:,} simulations)",
             fontweight="bold", pad=14, fontsize=13)
ax.set_xlabel("Win probability (%)")
ax.set_xlim(0, max(win_vals) * 1.18)
plt.tight_layout()
plt.savefig("outputs/win_probabilities.png", dpi=150, bbox_inches="tight")
plt.show()
```

## Visualisation 2 — Stacked progression chart

```python
top12 = list(results.items())[:12]
t_names  = [t for t,_ in top12]
win_v    = [s["win_pct"]   for _,s in top12]
final_v  = [s["final_pct"] for _,s in top12]
semi_v   = [s["semi_pct"]  for _,s in top12]
qf_v     = [s["qf_pct"]    for _,s in top12]

x = np.arange(len(t_names))
w = 0.18
fig, ax = plt.subplots(figsize=(14, 5))
ax.bar(x - 1.5*w, qf_v,    w, label="QF %",    color="#60a5fa", alpha=0.9)
ax.bar(x - 0.5*w, semi_v,  w, label="Semi %",  color="#34d399", alpha=0.9)
ax.bar(x + 0.5*w, final_v, w, label="Final %", color="#fbbf24", alpha=0.9)
ax.bar(x + 1.5*w, win_v,   w, label="Win %",   color="#f87171", alpha=0.9)

ax.set_xticks(x); ax.set_xticklabels(t_names, rotation=35, ha="right", fontsize=9)
ax.set_title(f"Tournament progression probabilities — top 12 ({N_SIMS:,} simulations)",
             fontweight="bold", pad=12)
ax.set_ylabel("Probability (%)")
ax.legend(fontsize=9)
ax.yaxis.grid(True, alpha=0.3); ax.set_axisbelow(True)
plt.tight_layout()
plt.savefig("outputs/progression_chart.png", dpi=150, bbox_inches="tight")
plt.show()
```

## Visualisation 3 — Group-by-group QF probability heatmap

```python
group_data = {}
for g, tbl in dict_table.items():
    group_data[g] = []
    for team in tbl["Team"].apply(clean):
        qf = results.get(team, {}).get("qf_pct", 0)
        group_data[g].append((team, qf))
    group_data[g].sort(key=lambda x: -x[1])

# Build matrix
group_names  = list(group_data.keys())
n_groups     = len(group_names)
max_per_group = max(len(v) for v in group_data.values())

matrix = np.zeros((n_groups, max_per_group))
labels = [["" for _ in range(max_per_group)] for _ in range(n_groups)]
for gi, grp in enumerate(group_names):
    for ti, (team, qf) in enumerate(group_data[grp]):
        matrix[gi, ti] = qf
        labels[gi][ti] = f"{team}\n{qf:.0f}%"

fig, ax = plt.subplots(figsize=(10, 9))
im = ax.imshow(matrix, cmap="YlOrRd", aspect="auto", vmin=0, vmax=matrix.max())
ax.set_xticks(range(max_per_group))
ax.set_xticklabels([f"Rank {i+1}" for i in range(max_per_group)], fontsize=9)
ax.set_yticks(range(n_groups))
ax.set_yticklabels(group_names, fontsize=9)
for gi in range(n_groups):
    for ti in range(max_per_group):
        if labels[gi][ti]:
            ax.text(ti, gi, labels[gi][ti], ha="center", va="center",
                    fontsize=7.5, color="black" if matrix[gi,ti] < matrix.max()*0.6 else "white")
ax.set_title("Quarter-final probability by group (% of simulations)",
             fontweight="bold", pad=12)
plt.colorbar(im, ax=ax, label="QF %")
plt.tight_layout()
plt.savefig("outputs/group_heatmap.png", dpi=150, bbox_inches="tight")
plt.show()
```

## Visualisation 4 — One sample bracket

```python
# Show a single simulation to illustrate the bracket
np.random.seed(18)
champ_s, tables_s, df_f_s, df_sf_s, df_qf_s = simulate_once(prob_cache)

print("=" * 50)
print("  SAMPLE TOURNAMENT SIMULATION")
print("=" * 50)

print("\n📋 Group winners:")
for g, tbl in tables_s.items():
    print(f"  {g}: 1st={tbl.iloc[0]['Team']}  2nd={tbl.iloc[1]['Team']}  3rd={tbl.iloc[2]['Team']}")

print("\n🏅 Quarter-finals:")
for _, r in df_qf_s.iterrows():
    print(f"  {r['Home']:20} vs {r['Away']:20}  →  {r['winner']}")

print("\n🏅 Semi-finals:")
for _, r in df_sf_s.iterrows():
    print(f"  {r['Home']:20} vs {r['Away']:20}  →  {r['winner']}")

print("\n🏅 Final:")
for _, r in df_f_s.iterrows():
    print(f"  {r['Home']:20} vs {r['Away']:20}  →  {r['winner']}")

print(f"\n🏆 CHAMPION: {champ_s}")
```
